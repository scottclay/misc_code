## Shannon Entropy

Entropy, in cyber security, is a measure of the randomness or diversity of a data-generating function. Data with full entropy is completely random and no meaningful patterns can be found. Low entropy data provides the ability or possibility to predict forthcoming generated values.

```python
#https://stackoverflow.com/questions/2257441/random-string-generation-with-upper-case-letters-and-digits
import string
import random
def id_generator(size=6, chars=string.ascii_uppercase + string.digits):
    return ''.join(random.choice(chars) for _ in range(size))
```

```python
str2 = ''
for i in range(0,len(str1.split())):
    N = random.randint(1, 10)
    str2+=id_generator(N)+' '
str2
```

```python
from collections import Counter
c1 = Counter(str1.split())
c2 = Counter(str2.split())
```

```python
#https://stackoverflow.com/questions/40496410/entropy-for-text-in-python
from math import log2
def shannon_entropy(counter):
    total_words = sum(counter.values())
    
    a = 0
    for i in counter.keys():
        a+=(counter[i]/total_words * log2(total_words / counter[i]))
    
    return a
```

```python
shannon_entropy(c1)
```

```python
shannon_entropy(c2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.figure(figsize=(7,7))

plt.xlim([0.85,2.15])
plt.ylim([0,15])
plt.xlabel("String", fontsize=16)
plt.ylabel("Shannon Entropy", fontsize=16)
plt.title("Shannon Entropy", fontsize=18)

plt.scatter(1, shannon_entropy(c1),label=r'Green Eggs and Ham', linewidth=2)
plt.scatter(2, shannon_entropy(c2),label=r'Random Strings', linewidth=2)
plt.legend(fontsize=14)

plt.gca().tick_params(axis='both', which='major', labelsize=14)
plt.gca().tick_params(axis='both', which='minor', labelsize=14)
```
